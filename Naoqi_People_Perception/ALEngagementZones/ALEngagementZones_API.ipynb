## Methods
___

> #### int ALEngagementZonesProxy::computeEngagementZone()

: 이 함수는 두가지 오버로드들을 가지고 있다.

1. ALEngagementZonesProxy::computeEngagementZone(x, y, z)는 FRAME_ROBOT에서 객체의 3D좌표를 사용한다.
2. ALEngagementZonesProxy::computeEngagementZone(xAngle,yAngle,distance,cameraPositionRobot)은 이미지에서 객체의 각좌표와 카메라와의 거리 및 FRAME_ROBOT의 카메라 위치를 사용한다.
    
    
___

> #### int ALEngagementZonesProxy::computeEngagementZone(const float& x, const float& y, const float& z)

#### Parameters : 
    1. x : FRAME_ROBOT에서의 Object의 x좌표값 
    2. y : FRAME_ROBOT에서의 Object의 y좌표값
    3. Z : FRAME_ROBOT에서의 Object의 z좌표값
    
#### Returns : 해당하는 Engagement 영역.

___

> #### int ALEngagementZonesProxy::computeEngagementZone(const float& xAngle, const float& yAngle, const float& distance, const AL::ALValue& cameraPositionRobot)

#### Parameters :
    1. xAngle : 이미지 안의 object의 수평 각도 좌표값(radian)
    2. yAngle : 이미지 안의 object의 수직 각도 좌표값(radian)
    3. distance : camera와 object 간의 거리
    4. cameraPositionRobot : Position6D로 FRAME_ROBOT에 있는 카메라 위치 (0은 top camera, 1은 bot camera)
    
#### Returns : 해당하는 Engagement 영역.

___

> #### float ALEngagementZonesProxy::getFirstLimitDistance()

#### Returns :
1. FirstDistance에 대한 현재 값. default값은 1.5m 이다. ALEngagementZones의 diagram을 참조하라.

___

> #### float ALEngagementZonesProxy::getLimitAngle()

#### Returns :
1. LimitAngle에 대한 현재 값. default값은 90도 이다. ALEngagementZones의 diagram을 참조하라.

___

> #### float ALEngagementZonesProxy::getSecondLimitDistance()

#### Returns :
1. SecondDistance에 대한 현재 값. default값은 2.5m 이다. ALEngagementZones의 diagram을 참조하라.

___

> #### bool ALEngagementZonesProxy::setFirstLimitDistance(const float& distance)

: FirstDistance의 값을 설정한다. 자세한 내용은 ALEngagementZones의 다이어그램을 참조하라. 이 메소드를 호출하면 EngagementZones/FirstLimitDistanceUpdated() 이벤트가 트리거 됩니다.

#### parameter : 
1. distance - 새로운 FirstDistance의 값(미터). 양수여야하며 SecondDistance의 현재 값보다 작아야한다.

#### Returns : 
1. 성공하면 true 출력

___

> #### bool ALEngagementZonesProxy::setLimitAngle(const float& angle)

: LimitAngle의 값을 설정한다. 자세한 내용은 ALEngagementZones의 다이어그램을 참조하라. 이 메소드를 호출하면 EngagementZones/LimitAngleUpdated() 이벤트가 트리거 됩니다.

#### parameter : 
1. distance - 새로운 LimitAngle의 값(도). 0과 180도 사이의 값이어야 한다.

#### Returns : 
1. 성공하면 true 출력

___

> #### bool ALEngagementZonesProxy::setSecondLimitDistance(const float& distance)

: SecondDistance의 값을 설정한다. 자세한 내용은 ALEngagementZones의 다이어그램을 참조하라. 이 메소드를 호출하면 EngagementZones/SecondLimitDistanceUpdated() 이벤트가 트리거 됩니다.

#### parameter : 
1. distance - 새로운 SecondDistance의 값(도). 양수여야하며 FirstDistance의 현재 값보다 커야한다.

#### Returns : 
1. 성공하면 true 출력

## ALMemory Keys
___

#### int PeoplePerception/Person/<ID>/EngagementZone
=> 해당 사람의 engagement 영역에 대한 ID(모를경우 0). <ID> 는 사람의 ID이다.(see ALPeoplePerception API).
    
#### AL::ALValue EngagementZones/PeopleInZone1
=> Zone 1에서 detected된 사람들의 ID에대한 list.

#### AL::ALValue EngagementZones/PeopleInZone2
=> Zone 2에서 detected된 사람들의 ID에대한 list.

#### AL::ALValue EngagementZones/PeopleInZone3
=> Zone 3에서 detected된 사람들의 ID에대한 list.

#### AL::ALValue EngagementZones/LastMovementsInZone1
=> Zone 1에서 detected된 움직임들의 list. 이때의 memory key는 MovementDetection/MovementInfo key와 같은 구조를 갖는다. ALMovementDetection에

#### AL::ALValue EngagementZones/LastMovementsInZone2
List of the movements detected in Zone 2. The memory key has the same structure as the key MovementDetection/MovementInfo, described in ALMovementDetection

#### AL::ALValue EngagementZones/LastMovementsInZone3
List of the movements detected in Zone 3. The memory key has the same structure as the key MovementDetection/MovementInfo, described in ALMovementDetection
