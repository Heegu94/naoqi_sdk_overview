#### bool ALFaceDetectionProxy::clearDatabase()

    database에 존재하는 모든 학습된 얼굴들을 제거한다.
        returns : 작동이 성공하면 true 출력

#### bool ALFaceDetectionProxy::forgetPerson(const std::string& name)

    지정된 사람에 해당되는 모든 학습된 얼굴들을 데이터 베이스에서 제거한다.

        Parameters:	name – The name of the person to forget
    
        Returns: true if the operation succeeded
        
___

#### AL::ALValue ALFaceDetectionProxy::getLearnedFacesList()

    학습한 각 얼굴의 이름이 포함된 list를 가져온다. 이 list의 크기는 항상 database에 존재하는 얼굴의 수와 같다.

        Returns:	List of names

#### float ALFaceDetectionProxy::getRecognitionConfidenceThreshold()

    face recognition에 사용되는 현재의 threshold 값을 반환한다.

    Returns:	Threshold value.

___

#### bool ALFaceDetectionProxy::importOldDatabase(const std::string& policy)

    Conversion tool. 오래된 database(NAOqi version < 1.22)을 새 형식으로 바꾼다. 이것은 한번만 실행하면 된다.

    module ALFaceDetection이 이미 database에 향해 있다면, 오래된 database를 import 할 수 있다. contents가 병합되고, 충돌하는 항복들(즉, 같은 이름을 갖는 얼굴들)이 지정된 policy에 따라 처리 된다.

        Parameters:	policy –
        
            import해온 항목의 이름이 기존 항목과 동일한 경우, policy를 병합한다.

            “overwrite"를 통해 기존 항목을 import해온 항목으로 바꿀 수 있다.

            “merge”하여 두 항목의 데이터를 병합한다. (동일한 사람과 관계가 없을경우, face recognition 은 실패할 수 있다),

            “keep”을 통해 기존 항목을 유지하고, import한 항목을 무시하도록 한다.
            
___

#### bool ALFaceDetectionProxy::learnFace(const std::string& name)

    새로운 얼굴을 학습하고, 특정 name 으로 database에 추가한다.

        Parameters: name – 학습할 사람의 이름
        
        Returns: true if the operation succeeded

___

#### bool ALFaceDetectionProxy::reLearnFace(const std::string& name)

    특정 사용자 기록에 새로운 Identification 사진을 추가한다.

    head pose가 다른 여러 이미지에서 사용자의 얼굴을 학습하면, recognition 결과를 향상시킬 수 있다. 한 사람당 최대 10개의 image들을 저장할 수 있으며, 그 다음에 새로운 이미지가 가장 오래된 이미지와 대체되는 방식(rolling_buffer_fashion)이 사용된다.

        Parameters:	name – 학습할 사람의 이름
        
        Returns: true if the operation succeeded
        
   > Notes : 만약 두명의 다른사람이 7초 이내에 인식되면 첫번째 사람만 다시 학습 할 수 있다.
___

#### void ALFaceDetectionProxy::setRecognitionConfidenceThreshold(float threshold)

    face recognition을 위한 새로운 threshold를 설정한다. 이 값은 0과 1 사이에 있어야하며, (default = 0.4)
    
        Parameters:	threshold – new threshold value.

for more details about threshold : (http://doc.aldebaran.com/2-4/naoqi/peopleperception/alfacedetection.html#threshold)

___

#### void ALFaceDetectionProxy::setRecognitionEnabled(const bool& enable)

    face recognition process를 enables/disables 한다. disable인 경우, 남아있는 face recognition이 빨라진다. (default : enable)

    Parameters:	enable – true/false
___

#### void ALFaceDetectionProxy::setTrackingEnabled(const bool& enable)

    face tracking을 enables/disables 한다. tracking을 enable 할 경우, 해당 사용자가 더이상 camera를 마주보지 않아도 얼굴을 tracking 할 수 있다. 그러나 이것은 잘 못된 detection으로 이어질 수 있다. active될 경우 한번에 한 얼굴만 감지된다. 

    Parameters:	enable – true/false
___

#### void ALFaceDetectionProxy::enableRecognition(const bool& enable)

    version 2.0 이후로 사용하지 않음 : Use ALFaceDetectionProxy::setRecognitionEnabled instead.
___

#### void ALFaceDetectionProxy::enableTracking(const bool& enable)¶

    version 2.0 이후로 사용하지 않음 : Use ALFaceDetectionProxy::setTrackingEnabled instead.
